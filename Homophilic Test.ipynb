```python
def calculate_CrossEdge(L):
    '''
    计算以某一个属性建图的CrossEdge数
    '''
    black_num = 0
    white_num = 0
    for i in L:
        if data['tag1'][i]==0:
            black_num +=1
        else:
            white_num +=1
    return black_num * white_num


def Statistic(attribute):
    '''
    计算以某一个属性建立的Unipartite的Edge数、CrossEdgeFraction
    '''
    X = data[data[attribute].notnull()][attribute]
    dic = {}
    for i in X.index:
        try:
            dic[X.loc[i]].append(i)
        except:
            dic[X.loc[i]] = []
            dic[X.loc[i]].append(i)
    #计算Cross_Edge_Num
    Cross_Edge_Num = 0
    for value in dic.values():
        n = calculate_CrossEdge(value)
        Cross_Edge_Num += n
    print 'Cross Edge Num:',Cross_Edge_Num
    #计算Edge_Num
    Edge_Num = 0
    for value in dic.values():
        if len(value)>1:
            Edge_Num += len(value)*(len(value)-1)/2
            
    print 'Edge Num:',Edge_Num
    print 'Cross Edge Fraction:',float(Cross_Edge_Num)/Edge_Num
    
```
